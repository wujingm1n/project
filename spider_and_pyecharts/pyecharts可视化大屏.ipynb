```python
import numpy as np
import pandas as pd
```

# 数据预处理

```python
df = pd.read_excel('hous_info.xlsx')
```

```python
df.info()
```

```python
df.head()
```

- 数据拆分

```python
df_temp = df['info'].str.split('|',expand=True)
df_temp
```

```python
df[['卧室数量', '厅室数量']] = df_temp.iloc[:,0].str.extract(pat='(\d+)室(\d+)厅', expand=True)
```

```python
df[['面积']] = df_temp.iloc[:,1].str.extract(pat=r'(.*?)平米', expand=True)
```

```python
df[['朝向']] = df_temp.iloc[:,2].str.extract(pat=r'([\u4e00-\u9fa5]+)', expand=True)
```

```python
df[['装修']] = pd.DataFrame(df_temp.iloc[:,3])
```

```python
df[['楼层', '总楼层']] = df_temp.iloc[:,4].str.extract(pat=r'(.*?)楼层\(共(\d+)层\)',expand=True)
```

```python
df[['建造年份']] = df_temp.iloc[:,5].str.extract(pat=r'(\d+)年', expand=True)
```

```python
df[['结构']] = pd.DataFrame(df_temp.iloc[:,6])
```

```python
df.head()
```

```python
df[['单价']] = df.loc[:,'unit_price'].str.extract(pat=r'(.*?)元')
```

```python
df['单价'] = df[['单价']].replace(to_replace='[,]',value='',regex=True)
```

```python
df[['总价']] = pd.DataFrame(df.loc[:,'price'].astype(np.int64))
```

```python
df.columns
```

```python
df.drop(labels=['info', 'info2', 'unit_price','price'], axis=1, inplace=True)
```

```python
dct = {'name': '小区',
       'addr': '区域'}
df.rename(mapper=dct, axis=1, inplace=True)
```

```python
df.info()
```

- 类型转换

```python
df['卧室数量'] = pd.to_numeric(df['卧室数量'])
df['厅室数量'] = pd.to_numeric(df['厅室数量'])
df['面积'] = pd.to_numeric(df['面积'])
df['总楼层'] = pd.to_numeric(df['总楼层'],downcast="signed")
df['建造年份'] = pd.to_numeric(df['建造年份'],downcast="signed")
df['单价'] = pd.to_numeric(df['单价'])
```

```python
df.info()
```

- 空值填充

```python
df.楼层.fillna(method='pad', inplace=True)
df.建造年份.fillna(df.建造年份.mode()[0], inplace=True)
df.结构.fillna(method='pad', inplace=True)
df.总楼层.fillna(df.总楼层.mode()[0], inplace=True)
```

```python
df.info()
```

```python
df.to_excel('hous_data.xlsx')
```

# 制作可视化大屏

```python
from pyecharts.charts import *
from pyecharts.components import Table
from pyecharts import options as opts
from pyecharts.commons.utils import JsCode
from pyecharts.globals import ThemeType
```

```python
df.head()
```

- 柱形图

```python
tb = (df.groupby('区域').agg({'单价': 'mean'}).sort_values(by='单价', ascending=False).reset_index())
tb['单价'] = tb['单价'].astype(int)
x, y = list(tb['区域']), list(tb['单价'])
# 以区域分组聚合单价的平均值
# 将单价舍去小数点，重设索引并降息排序
# 转为列表给x，y传参数
bar = Bar(init_opts=opts.InitOpts(bg_color='#080b30',  # 设置背景颜色
            theme='dark', # 设置主题
                                 ))
bar.add_xaxis(x)
bar.add_yaxis(series_name='房价',y_axis=y,
              label_opts=opts.LabelOpts(is_show=True, position='top',color='#fff'),
              markline_opts=opts.MarkLineOpts(data=[{'type': 'average','name': '平均值'}])
#              ,category_gap="70%" # 柱子宽度设置
             ) 
            
#设置y轴名称，给y轴传参数
bar.set_global_opts(title_opts=opts.TitleOpts(title='深圳各地段', subtitle="二手房交易均价"),
                    datazoom_opts=opts.DataZoomOpts(),
                    graphic_opts=opts.GraphicText(graphic_item=opts.GraphicItem(right='20%', top='10%'),
                    graphic_textstyle_opts=opts.GraphicTextStyleOpts(
                    text='香蜜湖,深圳湾,科技园,百花破十万!!',
                    font='bolder 20px Microsoft YaHei',
                    graphic_basicstyle_opts=opts.GraphicBasicStyleOpts(fill='white')
            )

))
bar.set_series_opts(itemstyle_opts={  
            "normal": {
                "color": JsCode(
                    """new echarts.graphic.LinearGradient(0, 0, 0, 1, [{
                        offset: 0,color: 'rgba(0, 244, 255, 1)'}
                        ,{offset: 1,color: 'rgba(0, 77, 167, 1)'}], false)
                    """
                ),       # 调整柱子颜色渐变
                'shadowBlur': 10,   # 光影大小
                "barBorderRadius": [100, 100, 100, 100],  # 调整柱子圆角弧度
                "shadowColor": "#0EEEF9", # 调整阴影颜色
                'shadowOffsetY': 2,
                'shadowOffsetX': 2,  # 偏移量
            }
        }
                   )
bar.render_notebook()
```

- 词云图

```python
import jieba
import jieba.analyse as anls
```

```python
#词云图
strings = ' '.join(str(x) for x in df['区域'])
jieba.analyse.set_stop_words('del_words.txt')
words = anls.extract_tags(
    strings,
    topK=50,
    withWeight=True,
    allowPOS=['ns', 'n', 'vn', 'v', 'nr'])
wc = WordCloud(init_opts=opts.InitOpts(bg_color='#080b30'))
wc.add( data_pair=words, word_size_range=[10, 100],series_name='高频词',
        shape='diamond',)
wc.set_global_opts(title_opts=opts.TitleOpts(title='深圳各地段', subtitle="二手房房源数量",title_textstyle_opts=dict(color='white')) )
wc.render_notebook()
```

- 箱型图

```python
x_data = df['区域'].unique().tolist()
y_data = [df[df['区域'] == x]['单价'].values.tolist() for x in x_data]
box = Boxplot(init_opts=opts.InitOpts(bg_color='#080b30',theme='dark'))
box.add_xaxis(x_data)
box.add_yaxis('房价', y_axis=box.prepare_data(y_data))                               
box.set_global_opts( title_opts=opts.TitleOpts(title='深圳各地段',subtitle= '箱型图'),
    datazoom_opts=opts.DataZoomOpts(),
                   )

box.set_series_opts(itemstyle_opts={  
            "normal": {
                "color": JsCode(
                    """new echarts.graphic.LinearGradient(0, 0, 0, 1, [{
                        offset: 0,color: 'rgba(0, 244, 255, 1)'}
                        ,{offset: 1,color: 'rgba(0, 77, 167, 1)'}], false)
                    """
                ),     
                'shadowBlur': 10,   # 光影大小
                "shadowColor": "#0EEEF9", # 调整阴影颜色
                'shadowOffsetY': 1,
                'shadowOffsetX': 1,  # 偏移量
            }
        }
                   )
box.render_notebook()
```

 - 散点图

```python
tb5 = df[['面积', '单价']].sort_values(by=['面积', '单价'], ascending=True)
x_sc, y_sc = list(tb5['面积']), list(tb5['单价'])
scatter = Scatter(init_opts=opts.InitOpts(bg_color='#080b30',theme='dark'))
scatter.add_xaxis(x_sc)
scatter.add_yaxis('', y_sc, label_opts=opts.LabelOpts(is_show=False))
scatter.set_global_opts(
    title_opts=opts.TitleOpts(title='面积与单价关系分布', subtitle = '散点图'),
    xaxis_opts=opts.AxisOpts(type_='value', name='面积', min_=0),
    yaxis_opts=opts.AxisOpts(type_='value', name='单价', min_=0),
    visualmap_opts=opts.VisualMapOpts(max_=max(y_sc), min_=min(y_sc)), 
)

scatter.render_notebook()
```

- 饼图

```python
tb = df.groupby('装修') \
    .agg(数量=('装修', 'count')) \
    .sort_values(by='数量', ascending=0) \
    .reset_index()
data = tb.values.tolist()
pie = Pie(init_opts=opts.InitOpts(bg_color='#080b30',theme='dark'))
pie.add(series_name='装修情况',
        data_pair=data,
        label_opts=opts.LabelOpts(formatter="{d}%")) #控制标签百分比显示
pie.set_global_opts(title_opts=opts.TitleOpts(title='装修情况',subtitle= '百分比',),
                    #控制图例位置
                    legend_opts=opts.LegendOpts(pos_left='left',
                                                pos_top='middle',
                                                orient='vertical'))

pie.render_notebook()
```

- 矩形树图

```python
tb = df.groupby('建造年份').agg(房源数量=('小区', 'count')).reset_index()
children = [{'name': x, 'value': y} for x, y in zip(tb['建造年份'], tb['房源数量'])]
data = [{'name': '年份', 'children': children}]
tm = TreeMap(init_opts=opts.InitOpts(bg_color='#080b30',theme='dark'))
tm.add(
    '房源数量',
    data,
    label_opts=opts.LabelOpts(
        # 水平排列居中
        horizontal_align='center',
        # 垂直排列靠上
        vertical_align='top',
        formatter='{b}: {c}'
    )
)
tm.set_global_opts(
    title_opts=opts.TitleOpts(title='各年份房源数量'),
    visualmap_opts=opts.VisualMapOpts(max_=max(tb['房源数量']),)
)
tm.render_notebook()
```

- 组件大屏

```python
# 大标题
table = Table()
v_title='深圳二手房数据'
table_color = '#080b30'
table.add(headers=[v_title], rows=[], attributes={
"align": "center",
"border": False,
"padding": "2px",
"style": "background:{}; width:1350px; height:50px; font-size:25px; color:#C0C0C0;".format(table_color)
})
table.render_notebook()
```

```python
# 先生成配置文件，在文件中进行配置之后保存，layout=Page.DraggablePageLayout为拖拽方式
page = Page(
     page_title = '深圳二手房',
    layout=Page.DraggablePageLayout)
page.add(table,bar,pie,wc,scatter,box,tm)
# 目的：通过拖放调整页面布局,并生成配置文件json
page.render('test.html')
# 在test.html中保存之后会生成一个配置文件
```

```python
# 读取配置文件后重新保存， source为原文件路径，cfg_file为配置文件路径，dest为最后保存的路径
page.save_resize_html(
    source='test.html',
    cfg_file='chart_config.json',
    dest='可视化大屏.html'
)
```

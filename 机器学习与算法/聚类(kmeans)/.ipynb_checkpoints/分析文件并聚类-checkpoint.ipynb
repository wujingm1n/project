# 分析文件

找出最佳聚类数目，并画出聚类的中心点


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.cluster import KMeans, DBSCAN
```

```python
# 导入数据
from scipy.io import loadmat
```

```python
data = loadmat('./data/ex7data2.mat')
```

```python
data
```

```python
X = data['X'].copy()
X.shape
```

## KMeans

```python
#  使用轮廓系数
from sklearn.metrics import silhouette_score
import warnings
warnings.filterwarnings('ignore')
```

```python
for k in range(2,10):
    kmeans = KMeans(k)
    labels = kmeans.fit(X).labels_
    score = silhouette_score(X,labels)
    print(score)
```

- 根据对轮廓系数的观察分三类最佳


```python
kmeans = KMeans(3)
labels = kmeans.fit(X).labels_
# 聚类中心点
centers = kmeans.cluster_centers_
plt.figure(dpi=150)
plt.scatter(X[:,0],X[:,1],c=labels)
plt.scatter(centers[:,0],centers[:,1],c='r')
plt.show()
```

## DBSCAN聚类

```python
plt.figure(figsize=(3*4, 3*5))

for eps in range(1, 10):

    db = DBSCAN(eps=eps*0.1, min_samples=3)
    labels = db.fit(X).labels_
    
    axes = plt.subplot(3, 3, eps)
    axes.scatter(X[:, 0], X[:, 1], c=labels)
    axes.set_title(f'eps={eps}', fontsize=20)
```

 -   从图中可以看出， eps=6,7,8,9的时候，聚类是比较合适的

## 两者区别

### 使用DBSCAN

```python
from sklearn.datasets import make_circles
```

```python
data, target = make_circles(
                                n_samples=300, 
                                noise=0.08,  # 噪声
                                factor=0.3  # 中间点的离散程度
                           )

plt.scatter(data[:, 0], data[:, 1], c=target)
```

```python
# DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=3)
dbscan.fit(data)
```

```python
plt.scatter(data[:, 0], data[:, 1], c=dbscan.labels_)
```

### 使用KMeans 查看效果区别

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)
```

```python
plt.scatter(data[:, 0], data[:, 1], c=kmeans.labels_)
```

 - KMeans并不适合这种数据集

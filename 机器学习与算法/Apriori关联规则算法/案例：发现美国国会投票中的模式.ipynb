# 前言
- 这⾥我们使⽤的是1984年美国国会投票的数据，这个数据集中包含了每个美国众议院议员对CQA确定的16个主要选票的投票。CQA列出了九种不同类型的投票：投票赞成、配对赞成和宣布赞成（这三个简化为yea)，投票反对、配对反对、宣布反对（这三个简化为nay）, 投票、为避免利益冲突⽽投票、弃权或以其他⽅式表明⽴场（这三个简化为未知意向）
- 数据集来源于UCI数据库：http://archive.ics.uci.edu/ml/datasets/Congressional+Voting+Records

- 这些强规则是在⽀持度为50%的情况下挖掘出来的，也就是说，这些规则⾄少在50%以上的记录中出现过。对于{y-预算决议案，n-医⽣费⽤冻结决议案}→{democrat}这条规则，在100%的情况下是成⽴的，也就是说，这是⼀个必然事件，只要他给预算决议案投了赞成票，并且医⽣费⽤冻结决议案投了反对票，那么这个⼈⼀定是⺠主党。
- 当设置⽀持度阈值为50%，置信度阈值为90%的时候，我们挖掘出来的强关联规则全部是跟⺠主党相关的，为什么会出现这样的情况呢?

```python
votes_['党派'].value_counts()
votes_['党派'].value_counts()/votes_.shape[0]
```
